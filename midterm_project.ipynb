````python
def read_stock_csv(filename):
    df = pd.read_csv(filename, header=[0,1])
    df.columns = [f"{price.strip().lower().replace(' ','_')}_{ticker.strip().lower()}" for price, ticker in df.columns]
    df = df.iloc[1:]
    df = df.rename(columns={df.columns[0]: 'Date'}) 
    df.index = range(1, len(df) + 1)

    for col in df.columns:
        if col != 'Date':
            df[col] = pd.to_numeric(df[col], errors='coerce')
    
    return df

universe = [
    'AAPL',
    'AMZN',
    'GOOG',
    'IBM',
    'META',
    'MSFT',
    'NFLX',
    'ORCL',
    'TSLA',
    'SAP',
]

stocks = {
    ticker: read_stock_csv(f"{ticker}_2018.csv") for ticker in universe
}

# for ticker, df in stocks.items():
#     print(ticker, df.shape)

# Unpacking the stocks into individual variables for easier access
(df_aapl, 
 df_amzn, 
 df_goog, 
 df_ibm,
 df_meta, 
 df_msft,
 df_nflx, 
 df_orcl, 
 df_tsla, 
 df_sap) = [stocks[t] for t in universe]

# df_aapl
df_ibm
````

````python
out = []

for ticker in universe:
    df = stocks[ticker]
    t = ticker.lower()
    out.append(df[f"adj_close_{t}"])
    out.append(df[f"close_{t}"])

    # Percentage change calculations
    pct_change = (
    (df[f'adj_close_{t}'] - df[f'adj_close_{t}'].shift(5)) 
    / df[f'adj_close_{t}'].shift(5) * 100
    ).round(2)
    out.append(pct_change.rename(f'5day_pct_change_{t}'))

    # Add dividend column
    dividend_col = f"dividends_{t}"
    if dividend_col in df.columns:
        out.append(df[dividend_col])
    else:
        out.append(pd.Series(0.0, index=df.index, name=dividend_col))

df_universe = pd.concat(out, axis=1) 
df_universe.insert(0, 'Date', stocks[universe[0]]['Date'].values)
df_universe.index = range(1, len(df_universe) + 1) 

df_universe
````

````python
pct_cols = [f"5day_pct_change_{t.lower()}" for t in universe]
close_cols = [f"close_{t.lower()}" for t in universe]
# print("Percent columns:\n", df_universe[pct_cols].dtypes)

# Average daily close
df_universe['avg_daily_close'] = df_universe[close_cols].mean(axis=1)

def get_max_5_stocks(row):
    values = pd.to_numeric(row[pct_cols], errors='coerce') 
    if values.isna().all():  
        return None
    top_5 = values.nlargest(5).index.tolist()
    stocks = [col.replace("5day_pct_change_", '').upper() for col in top_5]
    return ', '.join(stocks)

df_universe['max_5_stocks'] = df_universe.apply(get_max_5_stocks, axis=1)
# df_universe.style.format('{:,.2f}')

def get_min_5_stocks(row):
    values = pd.to_numeric(row[pct_cols], errors='coerce')
    if values.isna().all():
        return None
    bottom_5 = values.nsmallest(5).index.tolist()
    stocks = [col.replace("5day_pct_change_",'').upper() for col in bottom_5]
    return ', '.join(stocks)

df_universe['min_5_stocks'] = df_universe.apply(get_min_5_stocks, axis=1)

df_universe = df_universe.round(2)
df_universe
````

\end{equation*}

2. Your trading strategy is "5 days rebalancing of buying low". Here is how it works. You keep your portfolio unchanged until 5 days later on Jan 09 2018. Now you want to re-check the market and adjust your portfolio. You will compute the "Adj Close" price changes from Jan 02 to Jan 09, and find the 5 stocks whose "Adj Close" prices dropped the most in terms of percentage. You sell all current holdings on Jan 09 "Close" prices to convert your portfolio to all cash. Then immediately split your cash, including your cash account, to 5 equal parts to buy the 5 stocks that dropped the most from Jan 02 to Jan 09 on 'Adj Close' prices. You always buy the max shares of stock on the "Close" price and keep the rest cash in cash account. Now the portfolio should be different from 5 days ago. This operation is called "rebalancing".

    Keep in mind, the MTM will change every day, even when your portfolio holdings don't change, because the stock prices change.
